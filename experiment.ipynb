```python
import requests
from bs4 import BeautifulSoup
from io import BytesIO
from openpyxl import load_workbook

# URL of the page to scrape
base_url = "https://maps.vancouver.ca"
path_url = "/server/rest/services/VanMapViewer/Traffic_and_Transportation/MapServer/217/9852/attachments/"
session = requests.Session()

# Make a request to the URL
response = session.get(base_url+path_url)
response.raise_for_status()  # Ensure the request was successful

# Parse the HTML content
soup_link = BeautifulSoup(response.text, 'html.parser')

# Find the <a> tag with href ending in .xlsx or .pdf
attachment_link_tag = soup_link.find('a', text=lambda x: '.xlsx' in x or '.pdf' in x)

if attachment_link_tag:
    attachment_url = attachment_link_tag['href']
    full_url = base_url + attachment_url
    file_name = attachment_link_tag.text

# Initialize a session object
# Make a GET request to fetch the .xlsx file content
response = session.get(full_url)
response.raise_for_status()  # Ensure the request was successful

# Use BytesIO to create a file-like object from the content
file_content = BytesIO(response.content)

# Load the workbook from the file content
wb = load_workbook(filename=file_content)

# Access the active worksheet or a specific one by name
ws = wb.active  # or wb['Sheet1'] if you know the sheet name

# Read specific cells
cell_value = ws['M107'].value  # Replace 'A1' with the cell you're interested in

print(cell_value)  # Print the value of the specific cell
```
